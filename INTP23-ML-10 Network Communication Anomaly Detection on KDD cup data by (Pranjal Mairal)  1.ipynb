```python
#importing Naive bayes classifier 
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt

NB = GaussianNB()

NB.fit(X_train, Y_train.values.ravel())

Y_test_pred1 = NB.predict(X_test)


print("Train score is:", NB.score(X_train, Y_train))
print("Test score is:",NB.score(X_test,Y_test))
```

```python
#importing the libraries for artificial neural network
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
#creating a sequential model by defining the internal layers and giving the output and input dimension size
def fun():
    model = Sequential()
    
    #here 32 is output dimension
    model.add(Dense(32,input_dim =32,activation = 'relu',kernel_initializer='random_uniform'))
    
    #in next layer we do not specify the input_dim as the model is sequential so output of previous layer is input to next layer
    model.add(Dense(1,activation='sigmoid',kernel_initializer='random_uniform'))
    
    #5 classes-normal,dos,probe,r2l,u2r
    model.add(Dense(5,activation='softmax'))
    
    #loss is categorical_crossentropy which specifies that we have multiple classes
    
    model.compile(loss ='categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
    
    return model
```

```python
#building a model which will run for 100 epochs and as the data is large so we will feed the data in the natches of 1000 to the model
model7 = KerasClassifier(build_fn=fun,epochs=100,batch_size=1000)
```

```python
y_score = model3.fit(X_train, Y_train.values.ravel()).predict_proba(X_test)

from sklearn.preprocessing import LabelBinarizer

label_binarizer = LabelBinarizer().fit(Y_train)
y_onehot_test = label_binarizer.transform(Y_test)
y_onehot_test.shape  # (n_samples, n_classes)

label_binarizer.transform(["dos"])

class_of_interest = "dos"
class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)[0]
class_id

import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(
    y_onehot_test[:, class_id],
    y_score[:, class_id],
    name=f"{class_of_interest} vs the rest",
    color="darkorange",
)
plt.plot([0, 1], [0, 1], "k--", label="chance level (AUC = 0.5)")
plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("One-vs-Rest ROC curves:\dos vs (normal, probe, r2l, u2r)")
plt.legend()
plt.show()
```
